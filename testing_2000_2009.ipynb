```python
import pandas as pd
import numpy as np
import pyarrow.parquet as pq
import os
import yaml
import re
```

```python
PATH  = "src/agreste_2000_2009/"
file_list = [PATH + f for f in os.listdir(PATH) if f[-8:] in [f"{i}.csv" for i in range(2000, 2009)]]
```

```python
file_list
```

```python
dfs = []
for file in file_list:
    df = pd.read_csv(file, sep=';')
    dfs.append(df)
data = pd.concat(dfs, ignore_index=True)
```
